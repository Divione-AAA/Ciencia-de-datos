```python
import albumentations as A
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
#Carga de imagen
imagen = cv2.imread("532812.jpg")
img = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
```

```python
#Transformacion de imagenes
def transform(img):
    transform = A.Compose([
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        A.RandomRotate90(p=0.5),
        A.Rotate(limit=180, p=0.5),
        A.RandomBrightnessContrast(p=0.5),
    ])
    img = transform(image=img)['image']
    return img

ima = transform(img)
```

```python
#Plotearlo
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Augmentada")
plt.imshow(ima)
plt.axis("off")

plt.show()
```
