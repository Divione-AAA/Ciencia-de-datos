```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
x
```

```python
plt.plot(x,y)
plt.show
plt.xlabel("funcion de x")
plt.ylabel("funcion de y")
plt.title("Grafica de x en funcion de y")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Orientado a Objetos
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```
