```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
x
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Orientado a Objetos
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
# Grfica de dispersin
plt.scatter(x, y, color='green', marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Grfica de dispersin")
plt.show()
```

```python
# Grfica de barras
categorias = ['A', 'B', 'C', 'D']
valores = [10, 24, 36, 18]
plt.bar(categorias, valores, color='orange')
plt.xlabel("Categoras")
plt.ylabel("Valores")
plt.title("Grfica de barras")
plt.show()
```

```python
# Histograma
datos = np.random.randn(100)
plt.hist(datos, bins=15, color='purple', alpha=0.7)
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title("Histograma de datos aleatorios")
plt.show()
```

```python
# Mltiples lneas
plt.plot(x, y, label='x^2')
plt.plot(x, np.sqrt(x), label='sqrt(x)')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Varias funciones en una grfica")
plt.legend()
plt.show()
```
