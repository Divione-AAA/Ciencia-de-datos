```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Simulación de datos: inversión en publicidad vs. ventas
data = {
    'Publicidad (miles $)': [1, 2, 3, 4, 5, 6],
    'Ventas (miles $)': [3, 4, 2, 5, 6, 7]
}
df = pd.DataFrame(data)
```

```python
X = df[['Publicidad (miles $)']]  # variable independiente
y = df['Ventas (miles $)']       # variable dependiente
```

```python
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
print("Pendiente (m):", modelo.coef_[0])
print("Intercepto (b):", modelo.intercept_)
```

```python
nuevos_datos = pd.DataFrame({'Publicidad (miles $)': [7, 8, 9]})
predicciones = modelo.predict(nuevos_datos)
print(predicciones)
```

```python
# Gráfico de puntos + línea de regresión
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, modelo.predict(X), color='red', label='Línea de regresión')
plt.xlabel('Publicidad (miles $)')
plt.ylabel('Ventas (miles $)')
plt.title('Regresión Lineal')
plt.legend()
plt.grid(True)
plt.show()
```
