```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
texto = [
    "Me encanta aprender ciencia de datos",
]
stop_words_spanish = nltk.corpus.stopwords.words('spanish')
vectorizer = TfidfVectorizer(stop_words=stop_words_spanish)
X = vectorizer.fit_transform(texto)
```

```python
print(vectorizer.get_feature_names_out())
print(X.toarray())
```
