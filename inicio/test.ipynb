```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ejemplo
df = pd.DataFrame({
    "Día": ["Lun", "Mar", "Mié", "Jue", "Vie"],
    "Ventas": [100, 120, 90, 110, 95]
})

# Mostrar tabla
display(df)

# Graficar
df.plot(x="Día", y="Ventas", kind="bar", legend=False)
plt.title("Ventas por día")
plt.ylabel("Ventas")
plt.show()
```
